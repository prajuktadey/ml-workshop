# **Matplotlib**

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4,5,6,7])
y= np.array([10, 5,7,6,11,6,7])
```

```python
plt.scatter(x,y,marker= 'o', c='red')
plt.xlabel('X-->')
plt.ylabel('Y-->')
plt.title('X vs Y')
plt.show()
```

```python
y=2*x +5
plt.plot(x,y, "--")
plt.show()
```

```python
y=2*x +5
plt.plot(x,y, "o")
plt.show()
```

```python
#for a line of the form y=mx+c (y=1+x)
y = 2+3*x
plt.plot(x, y, '--')
```

```python
#for loop
for i in range (0,10,1): #takes three arguments, starts from 0, ends at 10, increments by 1
 if i%2==0:
   print(i, end=' ') #printing a space using end= ' '
 else:
   print(-1*i, end= ' ')
      
```

```python
# while loop
n= int(input("Enter  number: "))
rev= 0
while(n>0):
  rev= (rev*10)+ (n%10)
  n //=10 #//: integer divide operator, /: floor division
  #52/10: 5.2, 52//10: 5
print(rev)
```

```python
# lists
l = [1,2,3,4,5]
print("The list is: ", l)
print("The even elements are: ")
for i in range(len(l)):
    if l[i] % 2 == 0:
        print(l[i])
    else:
        l[i] = l[i] + 1
print("new list: ", l)
```
